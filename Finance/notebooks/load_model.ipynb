```python
import pandas as pd
import numpy as np
import joblib
import os
```

```python
df = pd.read_csv("../data/processed/credit_data_cleaned.csv")
print(df.shape)
df.head()
```

```python
model_dir = "../models"

rf = joblib.load(f"{model_dir}/random_forest_model.pkl")
print("Random Forest model loaded")
```

```python
X = df.drop(columns=["SeriousDlqin2yrs"])
y = df["SeriousDlqin2yrs"]

feature_columns = list(X.columns)
debt_ratio_median = df["DebtRatio"].median()

print("Feature columns:", feature_columns)
print("DebtRatio median:", debt_ratio_median)
```

```python
app_model_dir = "../app/model"
os.makedirs(app_model_dir, exist_ok=True)

joblib.dump(rf, f"{app_model_dir}/random_forest_model.pkl")
joblib.dump(feature_columns, f"{app_model_dir}/feature_columns.pkl")
joblib.dump(debt_ratio_median, f"{app_model_dir}/debt_ratio_median.pkl")

print("Model, feature columns, and debt ratio median exported to:", app_model_dir)
```

```python
model_check = joblib.load(f"{app_model_dir}/random_forest_model.pkl")
columns_check = joblib.load(f"{app_model_dir}/feature_columns.pkl")
median_check = joblib.load(f"{app_model_dir}/debt_ratio_median.pkl")

print("Model type:", type(model_check))
print("Number of features:", len(columns_check))
print("DebtRatio median:", median_check)
```
