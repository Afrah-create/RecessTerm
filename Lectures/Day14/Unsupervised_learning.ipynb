```python
image_names = [
    "dog_1.png", "dog_2.png", "dog_3.png", "dog_4.png", "dog_5.png", "dog_6.png",
    "cat_1.png", "cat_2.png", "cat_3.png", "cat_4.png", "cat_5.png", "cat_6.png",
    "car_1.png", "car_2.png", "car_3.png", "car_4.png", "car_5.png", "car_6.png",
    "mountain_1.png", "mountain_2.png", "mountain_3.png", "mountain_4.png", "mountain_5.png", "mountain_6.png",
    "unknown_1.png", "unknown_2.png", "unknown_3.png", "unknown_4.png", "unknown_5.png", "unknown_6.png", "Queen.jpeg"
]
```

```python
image_features = np.array([

# Dogs
[12,18],[14,20],[13,17],[15,19],[11,18],[16,20],

# Cats
[60,65],[62,67],[59,64],[63,66],[61,63],[60,68],

# Cars
[120,20],[118,18],[122,22],[119,19],[121,21],[123,20],

# Mountains
[200,180],[198,182],[201,179],
[203,181],[199,178],[202,183],[145,34],

# Unknown Images
[250,50],     # Flower
[180,20],     # Airplane
[170,100],    # Bicycle
[90,150],     # City
[230,90],     # Burger
[180,200]     # Beach

])
```

```python
# Create a K-Means clustering model with 4 clusters
kmeans = KMeans(n_clusters=4, random_state=42)

# Train the model on the image features
kmeans.fit(image_features)

# Cluster assignments for each image
cluster_assignments = kmeans.labels_
```

```python
# Display the cluster assignments for each image
for image_name, cluster in zip(image_names, cluster_assignments):
    print(f"Image: {image_name} is assigned to Cluster: {cluster}")

# Cluster centers
cluster_centers = kmeans.cluster_centers_
```

```python
# Reinforcement learning is a type of machine learning where an agent learns to 
# make decisions by interacting with an environment. The agent receives feedback 
# in the form of rewards or penalties based on its actions, and it aims to maximize 
# cumulative rewards over time. Common techniques include Q-learning, 
# Deep Q-Networks (DQN), and Policy Gradient methods. 
# Real-world applications of reinforcement learning include robotics, 
# game playing (e.g., AlphaGo), autonomous vehicles, and recommendation systems.

# Key points about reinforcement learning:
# 1. Agent and Environment: The agent interacts with the environment by taking 
# actions and receiving feedback.
# 2. Rewards: The agent receives rewards or penalties based on its actions, 
# guiding it toward optimal behavior.
# 3. Exploration vs. Exploitation: The agent must balance exploring new actions 
# to discover their effects and exploiting known actions that yield high rewards.
# 4. Markov Decision Process (MDP): Reinforcement learning problems are often 
#    modeled as MDPs,
#    where the agent's decisions depend on the current state and the expected 
# future rewards.
 # Applications: Reinforcement learning is used in various domains, including
# robotics, game playing, autonomous vehicles, and recommendation systems.
# Smart traffic control, personalized recommendations, and financial trading are some
# of the real-world applications of reinforcement learning.
# Fianace, Drone navigation, healthcare, and energy management are also areas 
# where reinforcement learning is being applied to optimize decision-making processes.
```

```python
# Example to demonstrate reinforcement learning concepts:
# Using an if statement to simulate a simple reinforcement learning scenario where an agent
# learns to choose between two actions based on rewards.
# Give reward to a robot for reaching the right destination and penalty for hitting an obstacle.

import random
# Enviroment
goal = 5
position = 0
# Output
print("Robot starts at position:", position)
while position != goal:
    # Agent chooses an action: move left (-1) or move right (+1)
    action = random.choice(['LEFT', 'RIGHT'])
    
    if action == 'RIGHT':
        position += 1
        print("Robot moves RIGHT to position:", position)
        if position == goal:
            print("Robot reached the goal! Reward: +10")
    else:
        position -= 1
        print("Robot moves LEFT to position:", position)
        if position < 0:
            print("Robot hit an obstacle! Penalty: -5")
            position = 0  # Reset position to 0 after hitting an obstacle
    # Reward and penalty feedback is given based on the robot's actions, 
    # guiding it toward the goal.
```
