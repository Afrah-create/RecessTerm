```python
image_names = [
    "dog_1.png", "dog_2.png", "dog_3.png", "dog_4.png", "dog_5.png", "dog_6.png",
    "cat_1.png", "cat_2.png", "cat_3.png", "cat_4.png", "cat_5.png", "cat_6.png",
    "car_1.png", "car_2.png", "car_3.png", "car_4.png", "car_5.png", "car_6.png",
    "mountain_1.png", "mountain_2.png", "mountain_3.png", "mountain_4.png", "mountain_5.png", "mountain_6.png",
    "unknown_1.png", "unknown_2.png", "unknown_3.png", "unknown_4.png", "unknown_5.png", "unknown_6.png", "Queen.jpeg"
]
```

```python
image_features = np.array([

# Dogs
[12,18],[14,20],[13,17],[15,19],[11,18],[16,20],

# Cats
[60,65],[62,67],[59,64],[63,66],[61,63],[60,68],

# Cars
[120,20],[118,18],[122,22],[119,19],[121,21],[123,20],

# Mountains
[200,180],[198,182],[201,179],
[203,181],[199,178],[202,183],[145,34],

# Unknown Images
[250,50],     # Flower
[180,20],     # Airplane
[170,100],    # Bicycle
[90,150],     # City
[230,90],     # Burger
[180,200]     # Beach

])
```

```python
# Create a K-Means clustering model with 4 clusters
kmeans = KMeans(n_clusters=4, random_state=42)

# Train the model on the image features
kmeans.fit(image_features)

# Cluster assignments for each image
cluster_assignments = kmeans.labels_
```

```python
# Display the cluster assignments for each image
for image_name, cluster in zip(image_names, cluster_assignments):
    print(f"Image: {image_name} is assigned to Cluster: {cluster}")

# Cluster centers
cluster_centers = kmeans.cluster_centers_
```
